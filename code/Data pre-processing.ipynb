# Data Importing

```python
import pandas
```

```python
# load bots and humans JSON files
df_bots = pandas.read_json('../data-raw/bot_activities.json')
df_humans = pandas.read_json('../data-raw/human_activities_original-3.json')
```

```python
activities = (
    # concatenate dataframes to one combined dataframe
    pandas.concat(
        [df_bots.assign(category='bot'), 
         df_humans.assign(category='human')], 
         ignore_index=True
    )
    [['category', 'contributor', 'repository', 'activity', 'date']]
)
activities['date'] = pandas.to_datetime(activities['date'])
activities
```

```python
activities.to_csv('../data/activities.csv', index=False)
```
